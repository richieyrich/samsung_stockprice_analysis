Samsung Stock Price Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Samsung Dataset.csv')
df
```

```python
df.info()
```

```python
df.nunique()
```

```python
df.describe()
```

change the date object to datetime format

```python

# change the date object to datetime format
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

check for null values

```python

# checking null values
df.isnull().sum()
```

plot a line chart of price and date

```python

# line chart of price and date
plt.figure(figsize=(16,7))
plt.plot( df['Date'], df['Close'])
plt.plot( df['Date'], df['Open'], c='r', alpha=0.5)
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Samsung Price')
plt.show()
```

since 2016, samsung has huge uptrend in the stocprice

analyse 2016 to 2024 chart

```python

# line chart of price and date
plt.figure(figsize=(16,7))
plt.plot( df['Date'], df['Close'])
plt.plot( df['Date'], df['High'], c='r', alpha=0.5, label = 'highest price')
plt.plot( df['Date'], df['Low'], c='g', alpha=0.5, label = 'lowest price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.xlim(pd.to_datetime('2016-01-01'), pd.to_datetime('2024-01-01'))
plt.title('Samsung Price')
plt.show()
```

plot a barplot showing the volume trading

```python

# bar chart of volume
plt.figure(figsize=(16,7))
sns.barplot(data= df, x='Date', y='Volume')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.title('Samsung stock Volume')
plt.show()
```

plot a histogram of closing price

```python
plt.figure(figsize=(16,7))
sns.histplot(data= df, x='Close', bins=20)
# plt.xlabel('Date')
# plt.ylabel('Close')
plt.title('Samsung stock closing price')
plt.show()
```

```python
plt.figure(figsize=(16,7))
sns.histplot(data= df, x='Volume', bins=20)
plt.title('Samsung stock Volume')
plt.show()
```

scatterplot on date and volumes

```python

plt.figure(figsize=(16,7))
sns.scatterplot( x= df['Date'], y= df['Volume'] )
plt.title('highest volume traded')
plt.show()
```

find the correlation between different columns to understand how they are related.

```python
corrltn = df.corr()
corrltn
```

```python
sns.heatmap(corrltn, annot=True, lw=0.5)
plt.title('correlation between different columns')
plt.show()
```
